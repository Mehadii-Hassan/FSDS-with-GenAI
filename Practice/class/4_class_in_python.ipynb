# Method overriding in OOP

```python
class Phone:
    def __init__(self, price, brand, camera):
        print("Inside Phone Constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def get_price(self):
        return self.__price

    def buy(self):
        print("Buying a phone inside phone constructor")

class SmartPhone(Phone):
    def buy(self):
        print("Buying a Phone inside SmartPhone constructor")
```

```python
s = SmartPhone(20000, "Apple", 3)
```

```python
s.buy()
```

# Super keyword in OOP

```python
class Phone:
    def __init__(self, price, brand, camera):
        print("Inside Phone Constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def get_price(self):
        return self.__price

    def buy(self):
        print("Buying a phone inside phone constructor")

class SmartPhone(Phone):
    def buy(self):
        print("Buying a Phone inside SmartPhone constructor")

        super().buy()
```

```python
s = SmartPhone(18000, "Android", 2)
```

```python
s.buy()
```

```python
class Phone:
    def __init__(self, price, brand, camera):
        print("Inside Phone Constructor")
        self.price = price
        self.brand = brand
        self.camera = camera

class SmartPhone(Phone):
    def __init__(self, price, brand, camera, os, ram):
        print("Inside SmartPhone constructor")
        super().__init__(price, brand, camera)
        self.os = os
        self.ram = ram
```

# Types of Inheritance in OOP

* Single 
* Multilevel
* Hierachical 
* Multiple

```python
# Single
class Phone:
    def __init__(self, price, brand, camera):
        print("Inside Phone Constructor")
        self.price = price
        self.brand = brand
        self.camera = camera

class SmartPhone(Phone):
    def __init__(self, price, brand, camera, os, ram):
        print("Inside SmartPhone constructor")
        super().__init__(price, brand, camera)
        self.os = os
        self.ram = ram
```

```python
# Multilevel

#GrandFather
class Product:
    def review(self):
        print("Product customer review")

#Father
class Phone(Product):
    def __init__(self, price, brand, camera):
        print("Inside Phone Constructor")
        self.price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("Buying a phone inside phone constructor")

#Child
class SmartPhone(Phone):
    pass


s = SmartPhone(20000, "Samsung", 3)
s.buy()
s.review()
```

```python
#Hierachical 

#Parent
class Phone:
    def __init__(self, price, brand, camera):
        print("Inside Phone Constructor")
        self.price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("Buying a phone inside phone constructor")


#child
class SmartPhone(Phone):
    pass

#child
class FeaturePhone(Phone):
    pass


s = SmartPhone(20000, "Samsung", 3)
s.buy()
```

```python
f = FeaturePhone(20000, "Samsung", 3)
f.buy()
```

```python
# Multiple
#Parent
class Phone:
    def __init__(self, price, brand, camera):
        print("Inside Phone Constructor")
        self.price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("Buying a phone inside phone constructor")

#Parent
class Product:
    def __init__(self):
        pass

    def review(self):
        print("Product customer review")

#child
class SmartPhone(Phone, Product):
    pass


s = SmartPhone(20000, "Samsung", 3)
s.buy()
s.review()
```
