# Method overriding in OOP

```python
class Phone:
    def __init__(self, price, brand, camera):
        print("Inside Phone Constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def get_price(self):
        return self.__price

    def buy(self):
        print("Buying a phone inside phone constructor")

class SmartPhone(Phone):
    def buy(self):
        print("Buying a Phone inside SmartPhone constructor")
```

```python
s = SmartPhone(20000, "Apple", 3)
```

```python
s.buy()
```

# Types of Inheritance in OOP

* Single 
* Multilevel
* Hierachical 
* Multiple

```python
# Single
class Phone:
    def __init__(self, price, brand, camera):
        print("Inside Phone Constructor")
        self.price = price
        self.brand = brand
        self.camera = camera

class SmartPhone(Phone):
    def __init__(self, price, brand, camera, os, ram):
        print("Inside SmartPhone constructor")
        super().__init__(price, brand, camera)
        self.os = os
        self.ram = ram
```

```python
# Multilevel

#GrandFather
class Product:
    def review(self):
        print("Product customer review")

#Father
class Phone(Product):
    def __init__(self, price, brand, camera):
        print("Inside Phone Constructor")
        self.price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("Buying a phone inside phone constructor")

#Child
class SmartPhone(Phone):
    pass


s = SmartPhone(20000, "Samsung", 3)
s.buy()
s.review()
```

```python
#Hierachical 

#Parent
class Phone:
    def __init__(self, price, brand, camera):
        print("Inside Phone Constructor")
        self.price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("Buying a phone inside phone constructor")


#child
class SmartPhone(Phone):
    pass

#child
class FeaturePhone(Phone):
    pass


s = SmartPhone(20000, "Samsung", 3)
s.buy()
```

```python
f = FeaturePhone(20000, "Samsung", 3)
f.buy()
```

```python
# Multiple
#Parent
class Phone:
    def __init__(self, price, brand, camera):
        print("Inside Phone Constructor")
        self.price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("Buying a phone inside phone constructor")

#Parent
class Product:
    def __init__(self):
        pass

    def review(self):
        print("Product customer review")

#child
class SmartPhone(Phone, Product):
    pass


s = SmartPhone(20000, "Samsung", 3)
s.buy()
s.review()
```

# Polymorphism in OOP

In python programming language, there are no polymorphism

```python
class Shape:
    def area(self, redius):
        print("This is first")
        return 3.4 * redius ** 2
    
    def area(self, l, b):
        print("This is second")
        return l * b
    

s = Shape()
s.area(2,3)
```

```python
class Shape:
    def area(self, redius):
        print("This is first")
        return 3.4 * redius ** 2
    
    def area(self, l, b):
        print("This is second")
        return l * b
    

s = Shape()
s.area(2)
```

```python
class Shape:
    def area(self, redius, l, b):
        if b == None:
            print("Radius")
            return 3.4 * redius ** 2
        else:
            print("Muliply")
            return l * b
    
s = Shape()
s.area(redius=8, l=45, b=None)
```

# Abstraction in OOP

```python
# Abstract Base Class (ABC)
from abc import ABC, abstractmethod

class BankServer(ABC):  #inherit with abstract
    def database(self):
        print("Connected to database")
    
    @abstractmethod #decorator
    def security(self):
        print("Security Passed")

class MobileApp:
    def mobile_login(self):
        print("Mobile app login successful")
    
    def security():
        print("Security Passed")
```

```python
mob = MobileApp()
```

```python
mob.mobile_login()
```
