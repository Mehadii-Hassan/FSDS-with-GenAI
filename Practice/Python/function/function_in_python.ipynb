```python
# Without parameter
def greeting():
    output = f"Hello Mehedi, Welcome to python"
    print(output) 

greeting()

# With parameter
def greeting(name):
    output = f"Hello {name}, Welcome to python"
    print(output)

greeting("Mehedi")
```

# Types of Arguments
* Default Argument
* Positional Argument
* Keyword Argument

```python
def power(a,b):
    return a**b

res = power(2,3)
print(res)
```

```python
def power(a=3, b=4): #default argument
    return a**b

res = power(2,2)
print(res)
```

```python
# Positional argument
def power(a, b): 
    return a**b

res = power(2,3) # 2 is store a, and 3 is store b
print(res)
```

```python
# Keyword argument
def power(a=3, b=4): 
    return a**b

res = power(a=2, b=5) # Keyword argument
print(res)
```

# *args and **kwargs

```python
# *args
def multiply(*args):
    product = 1
    for i in args:
        product = product * i
    return product

res = multiply(1,2,3,2,3,2,342,4)
print(res)
```

```python
# *kwargs
def display(**kwargs):
    for key,val in kwargs.items():
        print(f"{key} -> {val}")

display(india = "delhi", bangladesh = "Dhaka", srilanka = "colombo")
```

# Variable scope in function
* Global
* Local

```python
def g(y):
    print(x)
    print(x+1)

x = 5 #Global variable (variable declare outside of the function)
g(x)
print(x)
```

```python
x = "global awesome"

def my_function():
    x = "local fantastic" # This creates a new local variable 'x'
    print(f"Inside function: Python is {x}")

my_function() # Output: Inside function: Python is local fantastic
print(f"Outside function: Python is {x}") # Output: Outside function: Python is global awesome
```

```python
def f(y):
    x = 1 #local variable
    x += 1
    print(x)

x = 5 #global
f(x)
print(x)

# local variable first priority
# local variable can't accessable outside of the function
```

# Nested function

```python
def f():
    def g():
        print("Inside function g")
    print(g())
    
    print("Inside function f")

print(f())
```

```python
# recursion 
def f():
    def g():
        print("Inside function g")
        f()
    
    g()
    print("Inside function f")

f()
```

# Higher order functions

```python
L = [1,2,3,4,5]

def square(x):
    return x**2

def transform(f,L): #HOF
    output = []
    for i in L:
        output.append(f(i))
    print(output)

res = transform(square, L)
print(res)
```
