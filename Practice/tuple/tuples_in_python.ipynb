# Accessing Items
* Indexing
* Slicing

```python
t3 = (1, 2, 3, 4, 5)
print(t3[0])  # First element
print(t3[2])  # Third element
print(t3[-2]) # Second last element
print(t3[1:4]) # Slicing from index 1 to 3
print(t3[::2]) # Every second element
```

# Tuples operations
    * Arithmatic
    * Membership
    * Loop

```python
# Arithmetic operations
t1 = (1, 2, 3)
t2 = (4, 5, 6)
t3 = t1 + t2 # Concatenation
print(t3)

t1 * 4  # Repetition
print(t1 * 4)

# Membership
t3 = (1, 2, 3, 4, 5)
print(3 in t3)  # True
print(10 not in t3)  # True

# Looping through a tuple
t3 = (1, 2, 3, 4, 5)
for item in t3:
    print(item)
```

# Tuples method/function
* Len
* Min
* Max
* sorted
* sum

```python
t = (1, 2, 3, 4, 5)
print(len(t))  # Length of the tuple    
print(min(t))  # Minimum element
print(max(t))  # Maximum element
print(sorted(t))  # Sorted list of the tuple elements
print(sum(t))  # Sum of the elements
print(sorted(t, reverse=True))  # Sorted in descending order
```

```python
# count
t = (1, 2, 3, 2, 4, 2, 5)
print(t.count(2))  # Count of 2 in the tuple

# index
t = (1, 2, 3, 4, 5)
print(t.index(3))  # Index of first occurrence of 3
```

# Difference between Lists and tuples
* Syntax
* Mutability
* Speed
* Memory
* Built in functionality
* Error prone
* Usability

```python
import sys #sys mean system

L = list(range(1000))
T = tuple(range(1000))

print("List size:", sys.getsizeof(L), "bytes")
print("Tuple size:", sys.getsizeof(T), "bytes")
```

```python
a = (1, 2, 3)
b = a

a = a + (4,)
print(a)
print(b)
```

```python
# zip tuples
a = (1, 2, 3)
b = (4, 5, 6)

zipped = zip(a, b)
print(tuple(zipped))
```

# Tuple comprehension

```python
# Add 1 to 10 numbers to a list
L = []
for i in range(1, 11):
    L.append(i)
print(L)

# Using tuple comprehension (using generator expression)
T = tuple(i for i in range(1, 11))
print(T)
```

```python
# Scalar multiplication on a vector
v = [1, 2, 3]
s = -3

L = []

for i in v:
    result = i * s
    L.append(result)
print(L)

# Using tuple comprehension
T = tuple(i * s for i in v)
print(T)
```
