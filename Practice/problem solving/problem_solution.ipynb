## 1. Temperature Alert System
Write a program that takes today‚Äôs temperature as input.
If temperature > 35 ‚Üí print "Too Hot! Stay hydrated."
If temperature is between 20 and 35 ‚Üí print "Normal weather."
If temperature < 20 ‚Üí print "Cold weather, wear warm clothes."

```python
temperature = int(input("Enter today's temperature: "))

if temperature > 35:
    print("Too Hot! Stay hydrated.") 
elif temperature > 20 and temperature < 35:
    print("Normal Weater.")
elif temperature < 20:
    print("Cold weather, wear warm clothes.")
```

## 2. Movie Ticket Discount Checker
You ask the user for their age.
Age < 12 ‚Üí Child ticket discount (50% off)
12‚Äì18 ‚Üí Student discount (20% off)
18‚Äì60 ‚Üí No discount
Age > 60 ‚Üí Senior citizen discount (30% off)
Print which discount applies.

```python
age = int(input("Enter your age: "))
if age < 12:
    print("Child ticket discount (50% off)")
elif age >= 12 and age <= 18:
    print("Student discount (20% off)")
elif age >= 18 and age <= 60:
    print("No Discount.")
elif age > 60:
    print("Senior citizen discount (30% off)")
```

## Restaurant Bill with Discount
Problem:
Take total bill amount from the user:
‚óè If the bill is above 1000 ‚Üí apply 10% discount
‚óè Otherwise ‚Üí no discount
Then print the final bill amount.

## 

```python
bill_amount = int(input("Enter your total bill amount: "))

if bill_amount > 1000:
    discount = bill_amount * 0.1
    bill_amount = bill_amount - discount 
    print(f"You get 10% discount, your total bill is : {bill_amount}")
else:
    print(f"No discount. your total bill is: {bill_amount}")
```

## Leap Year Checker
Problem:
Take a year as input and check:
‚óè If it is divisible by 400 ‚Üí Leap Year
‚óè Else if divisible by 4 but not by 100 ‚Üí Leap Year
‚óè Otherwise ‚Üí Not a Leap Year

```python
year = int(input("Enter a year: "))

if year % 400 == 0:
    print("Leap Year")
elif year % 4 == 0 and year % 100 != 0:
    print("Leap Year")
else:
    print("Not Leap Year")
```

## Electricity Bill Calculator
Problem:
Input number of units consumed and calculate bill:
‚óè Up to 100 units ‚Üí 5 BDT per unit
‚óè 101‚Äì200 units ‚Üí 7 BDT per unit
‚óè Above 200 units ‚Üí 10 BDT per unit
Print total payable amount.

```python
units = int(input("Enter your Unit: "))

if units <= 100:
    units = units * 5
    print(f"Your total payable amount : {units}")
elif units >= 101 and units <= 200:
    units = units * 7
    print(f"Your total payable amount : {units}")
else:
    units = units * 10
    print(f"Your total payable amount : {units}")
```

## Email Validator (String Filtering + Loops)
üíº Real-world use: Used in signup forms or data cleaning tasks.

Example:
Input: hello@inceptionbd.com

## Count Word Frequency in Customer Reviews
üíº Real-world use: Analyzing reviews or comments in text analytics.
Task:
Given a review text, count how many times each word appears (ignore case).
Example:
Input: "This product is good and this service is excellent"
Output: {'this': 2, 'product': 1, 'is': 2, 'good': 1, 'and': 1, 'service': 1, 'excellent': 1}

```python
text = input("Enter your review: ").lower()
words = text.split()

d = {}
for i in words:
    if i not in d:
        d[i] = 1
    else:
        d[i] = d[i] + 1
print(d)

```
