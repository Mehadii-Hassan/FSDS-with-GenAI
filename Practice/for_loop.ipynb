```python
for i in "Mehedi":
    print(i, end=" ")
```

```python
for i in range(1, 5):
    for j in range(1, 5):
        print(i, j)
```

```python
row = int(input("Enter the number of rows"))

for i in range(1, row + 1):
    for j in range(1, i + 1):
        print("*", end=" ")

    print()
```

# Loop control statements
break,
continue.
pass


```python
for i in range(1, 11):
    if i == 5:
        break
    print(i)
```

```python
name = ["Mehedi", "Hasan", "Rony", "Khan"]

for i in name:
    if i == "Rony":
        print(i)
        break
```

```python
for i in range(1, 11):
    if i == 5:
        continue
    print(i)
```

```python
for i in range(1, 11):
    pass

```
