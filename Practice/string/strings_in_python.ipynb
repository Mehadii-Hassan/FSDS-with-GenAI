# Operations on String
    1. Arithmatic Operation
    2. Relational
    3. Logical
    4. Loops
    5. Membership

```python
#Arithmatic Operation
print("Hi" + " " + "Everyone")  #Concatenation
print("Hi" * 30)  #Repetition
```

```python
#Relational Operation

print("Hello" == "Hello")  #True
print("Hello" != "hello")  #True
print("Hello" > "hello")   #False
print("Hello" < "hello")   #True
print("Hello" >= "Hello")  #True

print("Hello" == "Hello")  #True
print("Hello" != "world")  #True
print("Hello" > "Hi")   #False
print("Hello" < "world")   #True
print("Hello" >= "Hello")  #True
```

```python
#Logical Operation

"Hello" and "World"  #Returns 'World'
```

```python
"Hello" or "World"   #Returns 'Hello'
```

```python
#Loops in string
for i in "Mehedi":
    print(i)
```

```python
#membership operation
"B" in "Bangladesh"  #True
"B" not in "Bangladesh"  #False
```

```python
s = "welcome to FSDS with Gen AI"
len(s)  #27
max(s)  #'w'
min(s)  #' '
sorted(s)  #[' ', ' ', ' ', ' ', 'A', 'D', 'F', 'G', 'I', 'S', 'S', 'c', 'e', 'e', 'e', 'h', 'i', 'l', 'm', 'o', 'o', 't', 't', 'w']
sorted(s, reverse=True)  #['w', 't', 't', 'o', 'o', 'm', 'l', 'i', 'h', 'e', 'e', 'e', 'c', 'S', 'S', 'F', 'G', 'I', 'A', ' ', ' ', ' ', ' ', 'D']
```

# String functions/methods
    1. Capitalize
    2. Title
    3. Upper
    4. Lower
    5. Swapcase

```python
s = "welcome to FSDS with Gen AI"

s.capitalize()  #'Welcome to fsds with gen ai'
s.title()  #'Welcome To Fsds With Gen Ai'
s.upper()  #'WELCOME TO FSDS WITH GEN AI'
s.lower()  #'welcome to fsds with gen ai'
s.swapcase()  #'WELCOME TO fsds WITH gEN ai'
```

```python
#palindrome 

word = str(input("Enter a word: "))

flag = True
for i in range(len(word)//2):
    if word[i] != word[len(word)-i -1]:
        flag = False
        break
    
if flag == True:
    print("Palindrome")
```
