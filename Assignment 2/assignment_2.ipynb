# Assignment 1: Email Validator (String Filtering + Loops)
üíº Real-world use: Used in signup forms or data cleaning tasks.
Task:
Check if a given email is valid ‚Äî must contain ‚Äú@‚Äù and end with ‚Äú.com‚Äù or ‚Äú.org‚Äù.
Example:
Input: hello@inceptionbd.com
Output: Valid Email

```python
Email = input("Enter your email address: ")

valid = True
special_characters = "!#$%^&*()-_=+[]{}|;:'\",<>/?`~"  

if " " in Email:
    valid = False

if Email.count("@") != 1:
    valid = False

if not (Email.endswith(".com") or Email.endswith(".org")):
    valid = False

if ".." in Email:
    valid = False


if valid:
    at_index = Email.index("@")

    first_part = Email[:at_index]
    second_part = Email[at_index+1:]

    # first part checks
    if len(first_part) == 0:
        valid = False
    elif first_part[0] in special_characters or first_part[-1] in special_characters:
        valid = False

    # second part checks
    if len(second_part) == 0:
        valid = False
    elif second_part[0] in special_characters:
        valid = False

# Final result
if valid:
    print("Valid Email")
else:
    print("Invalid Email")
```

# Assignment 2 : Count Word Frequency in Customer Reviews
üíº Real-world use: Analyzing reviews or comments in text analytics.
Task:
Given a review text, count how many times each word appears (ignore case).
Example:
Input: "This product is good and this service is excellent"
Output: {'this': 2, 'product': 1, 'is': 2, 'good': 1, 'and': 1, 'service': 1, 'excellent': 1}

```python
text_str = input("Enter a review or comment: ")
text_str2 = text_str.lower()
words = text_str2.split()

print("{", end="")

for i in range(len(words)):

    if words[i] in words[:i]:  
        continue

    count = words.count(words[i])
    is_last = True
    
    for j in range(i+1, len(words)):
        if words[j] not in words[:j]:
            is_last = False
            break

    if is_last:
        print(f"'{words[i]}': {count}", end="")
    else:
        print(f"'{words[i]}': {count}", end=", ")

print("}")
```

# Assignment 3 : Password Strength Checker
üíº Real-world use: Used in user authentication systems.
Task:
Check if a password is strong ‚Äî must have at least one uppercase, one lowercase, one digit, and one
special character.
Example:
Input: "Hello@123"
Output: Strong Password

```python
special_characters = "!@#$%^&*()-_=+[]{}|;:'\",.<>/?`~"

is_strong = False

while not is_strong:

    password = input("Enter a password: ")

    has_upper = False
    has_lower = False
    has_digit = False
    has_special = False

    for ch in password:
        if ch.isupper():
            has_upper = True
        elif ch.islower():
            has_lower = True
        elif ch.isdigit():
            has_digit = True
        elif ch in special_characters:
            has_special = True

    is_long = len(password) >= 8

    if has_upper and has_lower and has_digit and has_special and is_long:
        print("Strong Password")
        is_strong = True
    else:
        print("Weak Password")
        if not has_upper:
            print("- Add at least one uppercase letter (A-Z)")
        if not has_lower:
            print("- Add at least one lowercase letter (a-z)")
        if not has_digit:
            print("- Add at least one number (0-9)")
        if not has_special:
            print("- Add at least one special character (!,@,#,$,%, etc.)")
        if not is_long:
            print("- Password must be at least 8 characters long")
```

# Assignment 4 : Username Generator
üíº Real-world use: Automatically create usernames for users.
Task:
Take user‚Äôs full name and generate a short username.
Rules: first 3 letters of first name + last 3 letters of last name + random number (use loops only).
Example:
Input: "Mamun Malitha"
Output: mamitha57

```python
import random

full_name = input("Enter full name: ")

name_parts = []
word = ""

for ch in full_name:
    if ch == " ":
        if word != "":
            name_parts.append(word)
            word = ""
    else:
        word += ch

if word != "":
    name_parts.append(word)

first_name = name_parts[0]
last_name = name_parts[-1]

# First 3 letters
first_part = ""
for i in range(3):
    first_part += first_name[i]

# Last 3 letters
last_part = ""
for i in range(len(last_name)-3, len(last_name)):
    last_part += last_name[i]

# Random 2-digit number
rand_num = random.randint(10, 99)

user_name = first_part.lower() + last_part.lower() + str(rand_num)

print("Generated Username:", user_name)
```

# Assignment 5 : SMS Word Counter
üíº Real-world use: Used in messaging apps or social media post limit check.
Task:
Count how many characters and words are in a message, and tell if it exceeds 160 characters (SMS limit).
Example:
Input: "This is a demo message."
Output: Words = 5, Characters = 23, Status = Within Limit

```python
sms_word = input("Enter a demo message: ")

words = 0
in_word = False
characters = 0

for ch in sms_word:

    # Word count
    if ch != " " and not in_word:
        words += 1
        in_word = True
    elif ch == " ":
        in_word = False

    # Character count
    characters += 1

# SMS limit check
if characters <= 160:
    status = "Within Limit"
else:
    status = "Exceed Limit"

print(f"Words = {words}, Characters = {characters}, Status = {status}")
```

# Assignment 6 : Remove Duplicates from Contact List
üíº Real-world use: Data cleaning in CRM or user databases.
Task:
Given a list of names (comma separated), remove duplicates.
Example:
Input: "Rahim, Karim, Rahim, Sultana"
Output: "Rahim, Karim, Sultana"

```python
contact_str = "Rahim, Karim, Rahim, Rahim, Sultana"

contact_list = []
word = ""

# Manual split
for ch in contact_str:
    if ch == ",":
        if word.strip() != "":
            contact_list.append(word.strip())
        word = ""
    else:
        word += ch

if word.strip() != "":
    contact_list.append(word.strip())

# Remove duplicates manually
unique_list = []
for name in contact_list:
    found = False
    for u in unique_list:
        if u == name:
            found = True
            break
    if not found:
        unique_list.append(name)

# Build output
output = ""
for i in range(len(unique_list)):
    output += unique_list[i]
    if i != len(unique_list) - 1:
        output += ", "

print('Output: "' + output + '"')
```

# Assignment 7 : Detect Suspicious Words in Text (Moderation)
üíº Real-world use: Used in chat moderation or email filtering.
Task:
If any ‚Äúbanned‚Äù words like ['spam', 'scam', 'fake'] are found in a text, print ‚ÄúWarning‚Äù.
Example:
Input: "This is a scam offer"
Output: Warning: Suspicious content detected

```python
banned_words = ['spam', 'scam', 'fake']
found = False

text = input("Enter content: ")
lower_text = text.lower()

word = ""
text_list = []

# Split manually
for ch in lower_text:
    if ch != " ":
        word += ch
    else:
        if word != "":
            text_list.append(word)
            word = ""

# Append last word (important fix)
if word != "":
    text_list.append(word)

# Check banned words
for w in text_list:
    for bw in banned_words:
        if w == bw:
            found = True
            break
    if found:
        break

if found:
    print("Warning: Suspicious content detected.")
else:
    print("Safe: No banned words found")
```

# Assignment 8 : Invoice Code Formatter
üíº Real-world use: Formatting IDs or codes in finance or ERP systems.
Task:
Given an invoice number like 23, generate code in the format INV00023.
Example:
Input: 23
Output: INV00023

```python
number = int(input("Enter an invoice number: "))
number_to_str = str(number)

total_length = 5
zeros_needed = total_length - len(number_to_str)

zeros_str = ""
for _ in range(zeros_needed):
    zeros_str += "0"

code = "INV" + zeros_str + number_to_str
print("Generated Code:", code)
```

# Assignment 9 : Count Hashtags and Mentions (Social Media Analytics)
üíº Real-world use: Used in analyzing social media posts.
Task:
Count how many hashtags (#) and mentions (@) appear in a post.
Example:
Input: "Check out #AIWorkshop and follow @inceptionbd"
Output: Hashtags = 1, Mentions = 1

```python
post = input("Enter your post: ")

count_hashtag = 0
count_mention = 0

for ch in post:
    if ch == "#":
        count_hashtag += 1
    if ch == "@":
        count_mention += 1

print(f"Hashtags = {count_hashtag}, Mentions = {count_mention}")
```

# Assignment 10 : Word Reversal for Data Encryption
üíº Real-world use: Simple text obfuscation or playful encoding.
Task:
Reverse each word in a sentence while keeping word order the same.
Example:
Input: "AI is fun"
Output: "IA si nuf"

```python
my_string = input("Enter a sentence: ")

word = ""
reversed_list = []

for ch in my_string:
    if ch != " ":
        word += ch
    else:
        if word != "":
            reversed_list.append(word[::-1])
            word = ""

if word != "":
    reversed_list.append(word[::-1])

result = ""
for i in range(len(reversed_list)):
    result += reversed_list[i]
    if i != len(reversed_list) - 1:
        result += " "

print("Reversed Words Sentence:", result)
```
