```python
import logging
import os #for folder creating
```

```python
LOG_DIR = "logs" #folder name "logs"
LOG_FILE_NAME = "application.log" #file name

os.makedirs(LOG_DIR, exist_ok=True) 

log_path = os.path.join(LOG_DIR, LOG_FILE_NAME)

logging.basicConfig(
    filename= log_path,
    format= "[ %(asctime)s ] %(name)s - %(levelname)s - %(message)s",
    level= logging.INFO
)
```

```python
def profile(name):
    if type(name) == str:
        print("Welcome {name}")
    else:
        logging.info("Your input is not valid.")

person = profile(1234)
print(person)
```

```python
def devide_numbers(a,b):
    if b == 0:
        logging.info("Can't devide numbers with 0")
    else:
        return a / b

devide_numbers(4,0)
```
