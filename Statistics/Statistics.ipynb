# Measure Of Central Tendency
1. Mean
2. Median
3. Mode

```python
ages = [23,24,32,45,12,43,67,45,32,56,32]
```

# Calculation with numpy :
- Mean
- Median
- Mode

```python
# Calculate mean for ages
mean = np.mean(ages)
print(mean)

# Calculate median for ages
median = np.median(ages) 
print(median)

#Note : Numpy is not support is "Mode"
```

```python
ages = [23,24,32,45,12,43,67,45,32,56,32, 420] #has outlier  
```

```python
# Calculate mean for ages
mean = np.mean(ages)
print(mean)

# Calculate median for ages
median = np.median(ages) 
print(median)

#Note : Numpy is not support is "Mode" 
```

# Calculation with Statistics :
- Mean
- Median
- Mode

```python
# Calculate mean for ages
mean = statistics.mean(ages)
print(mean)

# Calculate median for ages
median = statistics.median(ages)
print(median)

# Calculate mode for ages
mode = statistics.mode(ages)
print(mode)
```

# Outlier Ditection

```python
ages = [23,24,32,45,12,43,67,45,32,56,32, 420] #has outlier
```

```python
sns.boxplot(ages)
```

# 5 Number Summary

```python
q1, q3 = np.percentile(ages,[25,75])
print(q1, q3)
```

```python
IQR = q3 - q1
lower_fence = q1 - 1.5*(IQR)
higher_fence = q3 + 1.5*(IQR)
print(lower_fence, higher_fence)
```

```python
for i in ages:
    if i > lower_fence and i < higher_fence:
        print(i)
```

# Measure Of Dispersion

1. Variance
2. Standard Deviation

```python
ages = [23,24,32,45,12,43,67,45,32,56,32]
```

```python
# Sample variance
statistics.variance(ages)
```

```python
# Sample Standard Deviation
statistics.stdev(ages)
```

```python
# Population variance
statistics.pvariance(ages)
```

```python
# sigma
math.sqrt(statistics.pvariance(ages)) 
```

# Log Normal Distribution, Power Law Distribution

```python
mu = 3.
sigma = 1.

s = np.random.lognormal(mu, sigma, 1000)
s
```

```python
def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample,kde=True)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist='norm',plot=pylab)
    plt.show()
```

```python
# create a normal distributed data
s=np.random.normal(0.5,0.2,1000)
```

```python
plot_data(s)
```
